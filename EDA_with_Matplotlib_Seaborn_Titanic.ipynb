## üß™ 5. Multivariate Analysis
Use hue, style, size to understand 3+ variables!

````python
# Hue & style in scatterplot
# hue: colour
# differentiate on the basis of class
sns.scatterplot(x='age', y='fare', hue='sex', style='class', data=df)
plt.title("Age vs Fare with Hue & Style")
plt.show()

# Pairplot
sns.pairplot(df[['age', 'fare', 'survived', 'pclass']].dropna(), hue='survived')
````

Histograms on the diagonal:

- Check the distribution of variables. For example:

- Age might show a skew towards younger passengers.

- Fare might have a right-skew, showing most passengers paid lower fares.

- Survived will be a binary distribution (0 = No, 1 = Yes).

- Pclass will show a distribution of the three classes.

Scatter plots:

- Look for trends or clusters:

- Age vs. Fare: You might see a wide range of ages for lower fares, and a few older passengers with higher fares.

- Age vs. Survived: You could see if younger or older passengers were more likely to survive.

- Pclass vs. Survived: This might show that higher-class passengers were more likely to survive.

## üß† 6. Correlation & Heatmaps
Check which features move together!

````python
# Correlation heatmap
corr = df.corr(numeric_only=True)
# correlation matrix : dependency on each other
# numeric only : leave the values that are not numbers

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
# annot=True	Adds the numeric values (correlation scores) inside each cell.
# cmap='coolwarm'	The color theme: blue for negative, red for positive correlations.
# fmt='.2f'	Format numbers to 2 decimal places, e.g., 0.89 instead of 0.89123.
plt.title("Correlation Matrix")
````

What's a correlation matrix?
It shows how strongly each numeric column is related to every other numeric column.

The values range from:

+1 ‚Üí Perfect positive correlation (as one increases, the other increases).

0 ‚Üí No correlation.

-1 ‚Üí Perfect negative correlation (as one increases, the other decreases).

Why numeric_only=True?
It tells Pandas to only include numeric columns (like age, fare, survived) and skip categorical/text columns (like sex, embarked, etc.).

## üé® 7. Matplotlib Customization
Titles, labels, figure sizes, saving plots

````python
# Multiple plots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
# figsize: wide*tall
# fig - assume white big sheet
# axs as the individual plots

# subplot- single subplot
# subplots- many subplots

sns.histplot(df['age'], ax=axs[0])
axs[0].set_title('Age Dist')

sns.histplot(df['fare'], ax=axs[1])
axs[1].set_title('Fare Dist')

plt.tight_layout()
plt.savefig("age_fare_dist.png")
````
